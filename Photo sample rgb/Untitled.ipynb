```python
import csv
import os
from Target_colors import target_colors_grn, target_colors_brn
import numpy as np
from PIL import Image
```

Directory info

```python
original_directory = os.getcwd()
os.chdir('../Pictures')
filename = '3010 C1 W-HW 9 water.JPG'
```

```python
img = Image.open(filename)
```

```python
size = img.size
size
```

```python
pix_val = img.getcolors(size[0]*size[1])
pix_val
```

$$\int_0^2 \sin(\pi x) dx = 0$$

```python
rowname = filename.split('.')[0]
row = [rowname]
```

```python
num_target_colors = len(target_colors_brn)
```

```python
for i in range(num_target_colors):
    print(i)
    for set in pix_val:
        if np.array_equal(set[1][:3],target_colors_brn[i]):
            row.append(set[0])
            break
    else:
        row.append(0)
        
```

```python
j_old = 0
for i in range(num_target_colors):
    print(i,j_old,pix_val[j_old][1])
    for j in range(j_old,len(pix_val)):
        if np.array_equal(pix_val[j][1][:3],target_colors_brn[i]):
            row.append(pix_val[j][0])
            j_old = j + 1
            break
    else:
        row.append(0)
        
```

```python
rows = [row]
os.chdir(original_directory)
with open('Target_Color_Brn_counts.csv', 'a', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerows(rows)
```
